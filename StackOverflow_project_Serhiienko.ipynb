Завдання 1. Підрахунок загальної кількості респондентів

```python
total_respondents = df['ResponseId'].nunique()
print ("Загальна кількість респондентів:", total_respondents)
```

Завдання 2. Аналіз повноти відповідей респондентів.

```python
schema_questions = set(df1['qname'].dropna().unique())
df_columns = set(df.columns)
questions = schema_questions.intersection(df_columns)
print("Кількість питань у перетині:", len(questions))
print("Приклади питань:", list(questions)[:10])
df_questions = df[list(questions)]
complete_responses = df_questions.dropna()
total_complete = complete_responses.shape[0]
print("Кількість респондентів, які відповіли на всі питання:", total_complete)
```

Завдання 3. Статистичний аналіз досвіду респондентів.

```python
df['WorkExp'].head()
workexp_clean = df['WorkExp'].dropna()
mean_exp = workexp_clean.mean()
median_exp = workexp_clean.median()
mode_exp = workexp_clean.mode()
print(f"Міри центральної тенденції для WorkExp:")
print(f"Середнє: {mean_exp:.2f} років")
print(f"Медіана: {median_exp:.2f} років")
print(f"Мода: {', '.join(map(str, mode_exp.values))} років")
```

Завдання 4. Аналіз віддаленої роботи.

```python
remote_count = df[df['RemoteWork'] == 'Remote'].shape[0]
print("Кількість респондентів, які працюють віддалено:", remote_count)
```

Завдання 5. Визначення популярності Python

```python
python_mask = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)
python_count = python_mask.sum()
total_respondets = df.shape[0]
python_percent = (python_count / total_respondents) * 100
print(f"Відсоток респондентів, які програмують на Python: {python_percent: .1f}%")
```

Завдання 6. Аналіз шляхів навчання програмуванню

```python
online_mask = df['LearnCode'].str.contains('Online Courses', case=False, na=False)
online_count = online_mask.sum()
print("Кількість респондентів, які навчалися через онлайн-курси:", online_count)
```

Завдання 7. Географічний аналіз компенсації Python-розробників.

```python
python_mask = df['LanguageHaveWorkedWith'] \
    .str.contains('Python', case=False, na=False)
python_devs = df[python_mask].copy()
python_devs_clean = python_devs \
    .dropna(subset=['ConvertedCompYearly']) \
    .copy()
country_comp_stats = (
    python_devs_clean
    .groupby('Country')['ConvertedCompYearly']
    .agg(
      Mean_Compensation='mean',
      Median_Compensation='median'
    )
    .round(2)
    .reset_index()
)
country_comp_stats.head(10)
```

Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

Завдання 9. Аналіз популярності Python по віковим категоріям

```python
df_age_clean = df.dropna(subset=['Age', 'LanguageHaveWorkedWith']).copy()
df_age_clean['Python_Dev'] = df_age_clean['LanguageHaveWorkedWith'] \
    .str.contains('Python', case=False, na=False)
age_group_stats = (
    df_age_clean
    .groupby('Age')['Python_Dev']
    .mean()
    .reset_index()
)
age_group_stats['Python_Percentage'] = (
    age_group_stats['Python_Dev'] * 100
).round(2)
age_group_stats = age_group_stats[['Age', 'Python_Percentage']]
age_group_stats.head(10)
```

Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
df_clean = df.dropna(subset=['ConvertedCompYearly','RemoteWork','Industry'])
percentile_75 = df_clean['ConvertedCompYearly'].quantile(0.75)
print("75-й перцентиль:", percentile_75)
high_paid_remote = df_clean[
    (df_clean['ConvertedCompYearly']>percentile_75)&
    (df_clean['RemoteWork'] == 'Remote')
]
industry_counts = high_paid_remote['Industry'].value_counts()
industry_counts.head(10)
```
